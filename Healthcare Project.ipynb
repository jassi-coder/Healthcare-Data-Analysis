```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import os

# -----------------------
# Create folders
# -----------------------
os.makedirs("outputs/charts", exist_ok=True)

# -----------------------
# Load Data
# -----------------------
df = pd.read_csv("Healthcare_dataset.csv")

# -----------------------
# Select Important Columns
# -----------------------
df = df[['Age', 'Gender', 'Medical Condition', 'Billing Amount', 'Admission Type']]

# -----------------------
# Create Age Groups
# -----------------------
df['Age Group'] = pd.cut(df['Age'],
                        bins=[0,18,35,50,65,100],
                        labels=['0-18','19-35','36-50','51-65','65+'])

# -----------------------
# Basic Analysis
# -----------------------
age_group = df['Age Group'].value_counts()
gender_count = df['Gender'].value_counts()
disease_count = df['Medical Condition'].value_counts()
admission_type = df['Admission Type'].value_counts()

billing_avg = df['Billing Amount'].mean()

# -----------------------
# Charts (IMPORTANT PART)
# -----------------------

# 1. Age Group Chart
age_group.plot(kind='bar')
plt.title("Age Group Distribution")
plt.xlabel("Age Group")
plt.ylabel("Patients")
plt.savefig("outputs/charts/age_group.png")
plt.show()
plt.close()

# 2. Disease Chart
disease_count.plot(kind='bar')
plt.title("Disease Distribution")
plt.xlabel("Disease")
plt.ylabel("Count")
plt.savefig("outputs/charts/disease.png")
plt.show()
plt.close()

# 3. Admission Type Pie Chart
admission_type.plot(kind='pie', autopct='%1.1f%%')
plt.title("Admission Type Distribution")
plt.savefig("outputs/charts/admission.png")
plt.show()
plt.close()

# 4. Age vs Billing Scatter
plt.scatter(df['Age'], df['Billing Amount'])
plt.title("Age vs Billing Amount")
plt.xlabel("Age")
plt.ylabel("Billing Amount")
plt.savefig("outputs/charts/age_vs_billing.png")
plt.show()
plt.close()

# -----------------------
# Report File
# -----------------------
with open("outputs/report.txt", "w") as f:
    f.write("Healthcare Data Analysis Report\n\n")
    f.write(f"Most common age group: {age_group.idxmax()}\n")
    f.write(f"Emergency cases (%): {(admission_type.get('Emergency',0)/len(df))*100:.2f}%\n")
    f.write(f"Most common disease: {disease_count.idxmax()}\n")
    f.write(f"Average Billing Amount: {billing_avg:.2f}\n")

```
